## **Projeto 3 - Anlise de dados - COVID-19**

O **SARS-CoV-2** (do ingls Severe Acute Respiratory Syndrome CoronaVirus-2), agente etiolgico da **COVID-19** (do ingls COronaVIrus Disease 2019), pertence  famlia dos **Coronavrus**, que so um grupo de vrus envelopados que possuem **RNA linear de fita simples sentido positivo**. O SARS-CoV-2 apresenta **quatro protenas** estruturais principais: a **glicoprotena de superfcie** (do ingls spike) (S), a **protena do envelope** (E), a **protena da matriz** (M) e a **protena do nucleocapsdeo** (N).

So uma grande **famlia de vrus** comuns em muitas espcies diferentes de animais, incluindo camelos, gado, gatos e morcegos. Raramente, os coronavrus que infectam animais podem infectar pessoas, como exemplo do **MERS-CoV e SARS-CoV**. Contudo, desde o incio da pandemia de **COVID-19**, anlises genticas do SARS-CoV-2 em vrios pases e em diferentes momentos revelaram que o vrus sofreu **diversas mutaes**. As mutaes so **eventos naturais da replicao viral**, sendo muito mais frequentes em vrus de genoma RNA. A maior parte das mutaes  **neutra**, ou seja, no fornece qualquer vantagem ou desvantagem para o vrus. Entretanto, um pequeno nmero de mutaes pode conferir **novas propriedades qumicas s protenas virais**, resultando em mudanas na forma como o vrus se comporta nas infeces.

Mais informaes sobre a pesquisa dos dados podem ser acessados na [RBAC](https://www.rbac.org.br/artigos/principais-variantes-do-sars-cov-2-notificadas-no-brasil/), revista brasileira de anlises clnicas. Os sintomas da COVID-19 podem variar de um **resfriado** a uma **sndrome gripal** - presena de um quadro respiratrio agudo caracterizado por pelo menos dois dos seguintes sintomas: **sensao febril ou febre** associada a dor de garganta, dor de cabea, tosse, coriza, ou at uma **pneumonia severa**. Sendo assim, os sintomas mais comuns so:

* Tosse;
* Febre;
* Coriza;
* Dor de garganta;
* Dificuldade para respirar;
* Perda de olfato (anosmia);
* Alterao do paladar (ageusia);
* Distrbios gastrintestinais (nuseas/vmitos/diarreia);
* Cansao (astenia);
* Diminuio do apetite (hiporexia);
* Dispnia ( falta de ar).

A **transmisso** acontece de uma pessoa doente para outra ou por contato prximo por meio de:

* Toque do aperto de mos contaminadas;
* Gotculas de saliva;
* Espirro;
* Tosse;
* Catarro;
* Objetos ou superfcies contaminadas, como celulares, mesas, talheres, maanetas, brinquedos, teclados de computador etc.

Criaremos **cpias** da base de dados para anlises mais assertivas: tanto para o nvel **mundial**, quanto para os continentes (**Europa, sia, frica, Amrica do Sul, e Oceania**), assim como para os **pases** em geral e uma para o **Brasil**. Exploraremos algumas informaes relevantes como a linha do tempo para o **total de casos**. Para os pases, o **total de mortes** e **total de mortes por 1 milho de pessoas**.

```
# 14 - Criando cpias da base de dados com filtros especficos
df_copia = df.copy()
df_mundo = df_copia[(df_copia.location == "World") & (df_copia.date <= "2024-01-01")]
df_europa = df_copia[(df_copia.continent == "Europe") & (df_copia.date <= "2024-01-01")]
df_asia = df_copia[(df_copia.continent == "Asia") & (df_copia.date <= "2024-01-01")]
df_africa = df_copia[(df_copia.continent == "Africa") & (df_copia.date <= "2024-01-01")]
df_america_sul = df_copia[(df_copia.continent == "South America") & (df_copia.date <= "2024-01-01")]
df_oceania = df_copia[(df_copia.continent == "Oceania") & (df_copia.date <= "2024-01-01")]
df_brasil = df_copia[(df_copia.location == "Brazil") & (df_copia.date <= "2024-01-01")]
df_paises = df_copia.query("location not in ['United Kingdom', 'Asia', 'Africa', 'Europe', 'European Union', 'International', 'North America', 'Oceania', 'South America', 'World']")
df_paises = df_paises.loc[df_paises.date <= "2024-01-01"]
```

Visualizando a linha do tempo de **total de casos no mundo**.

```
# 15 - Visualizando o total de casos por COVID no mundo
dia = df_mundo.groupby("date")["total_cases"].sum()
dia = pd.DataFrame(dia)
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize = (10 ,3), dpi = 150)
dia["total_cases"].plot(kind = "line", ax = ax, lw = 2.5)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: format(int(x), ",")))
ax.set_title("Linha do tempo do total de casos de COVID-19 no mundo at Jan/24", fontsize = 14, fontweight = "semibold", color = "grey")
ax.set_xlabel("")
plt.tight_layout()
```

Olhando para o grfico acima percebemos um **ntido crescimento** de casos prximo ao final de **2020** indo praticamente at **janeiro de 2022**. Em seguida, um **crescimento acentuado** entre **janeiro de 2022 e janeiro de 2023**, com uma relativa **estabilizao** a partir desta data.

Com essa informao, vamos analisar os tpicos: **total de mortes** e **total de mortes a cada 1 milho de pessoas** pelos **principais pases** e continentes.

```
# 16 - Visualizando o total de mortes por pas no mundo
ultima_data = df_paises['date'].max().strftime("%Y-%m-%d")
paises_ordenados = df_paises.loc[df_paises.date == ultima_data].sort_values(by = "total_deaths", ascending = False)
paises_ordenados = paises_ordenados[~paises_ordenados['location'].str.contains('income')][0:7]
paises_ordenados = paises_ordenados[["location", "total_deaths"]]

sns.set_style()
fig, ax = plt.subplots(figsize=(10, 3), dpi = 150)
barras = sns.barplot(x = "total_deaths", y = "location", data = paises_ordenados, palette = "cividis",
                     hue = "location", legend = False)

for i in barras.patches:
  ax.annotate(int(i.get_width()),
              xy = ((i.get_width() - (i.get_height() * 5000)),
                     i.get_x() + i.get_y() + (i.get_height() / 2)),
                     va = "center_baseline", color = "white", fontweight = "bold",
                     ha = "right")

for i in range(len(paises_ordenados)):
  pais = paises_ordenados.iloc[i, 0]
  valor = paises_ordenados.iloc[i, 1]
  ax.text(valor - valor + 5000, i, pais, fontsize = 10, fontweight = "bold", color = "white", va = "center")

plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
ax.text(0, -0.8, "Total de mortes por pas no mundo at Jan/24", fontsize = 14, color = "grey", fontweight = "bold")
ax.set_xlabel("")
ax.set_ylabel("")
plt.tight_layout()
```

Acima so os dados dos pases onde houve **maior nmero de mortes** por **COVID-19**. Em seguida, vamos visualizar o **total de mortes** a cada **1 milho** de pessoas por pases no **mundo**.

```
# 17 - Visualizando o total de mortes por pas no mundo a cada 1 milho de pessoas
ultima_data = df_paises["date"].max().strftime("%Y-%m-%d")
paises_ordenados = df_paises.loc[df_paises.date == ultima_data].sort_values(by = "total_deaths_per_million", ascending = False)
paises_ordenados = paises_ordenados[~paises_ordenados["location"].str.contains("income")][0:7]
paises_ordenados = paises_ordenados[["location", "total_deaths_per_million"]]

sns.set_style()
fig, ax = plt.subplots(figsize=(10, 3), dpi = 150)
barras = sns.barplot(x = "total_deaths_per_million", y = "location", data = paises_ordenados, palette = "copper",
                     hue = "location", legend = False)

for i in barras.patches:
  ax.annotate(int(i.get_width()),
              xy = ((i.get_width() - (i.get_height() * 30)),
                     i.get_x() + i.get_y() + (i.get_height() / 2)),
                     va = "center_baseline", color = "white", fontweight = "bold",
                     ha = "right")

for i in range(len(paises_ordenados)):
  pais = paises_ordenados.iloc[i, 0]
  valor = paises_ordenados.iloc[i, 1]
  ax.text(valor - valor + 30, i, pais, fontsize = 10, fontweight = "bold", color = "white", va = "center")

plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
ax.text(0, -0.8, "Total de mortes por pas no mundo a cada 1 milho de pessoas at Jan/24",
        fontsize = 14, color = "brown", fontweight = "bold")
ax.set_xlabel("")
ax.set_ylabel("")
plt.tight_layout()
```
