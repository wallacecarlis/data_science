## **Projeto 3 - Análise de dados - COVID-19**

O **SARS-CoV-2** (do inglês Severe Acute Respiratory Syndrome CoronaVirus-2), agente etiológico da **COVID-19** (do inglês COronaVIrus Disease 2019), pertence à família dos **Coronavírus**, que são um grupo de vírus envelopados que possuem **RNA linear de fita simples sentido positivo**. O SARS-CoV-2 apresenta **quatro proteínas** estruturais principais: a **glicoproteína de superfície** (do inglês spike) (S), a **proteína do envelope** (E), a **proteína da matriz** (M) e a **proteína do nucleocapsídeo** (N).

São uma grande **família de vírus** comuns em muitas espécies diferentes de animais, incluindo camelos, gado, gatos e morcegos. Raramente, os coronavírus que infectam animais podem infectar pessoas, como exemplo do **MERS-CoV e SARS-CoV**. Contudo, desde o início da pandemia de **COVID-19**, análises genéticas do SARS-CoV-2 em vários países e em diferentes momentos revelaram que o vírus sofreu **diversas mutações**. As mutações são **eventos naturais da replicação viral**, sendo muito mais frequentes em vírus de genoma RNA. A maior parte das mutações é **neutra**, ou seja, não fornece qualquer vantagem ou desvantagem para o vírus. Entretanto, um pequeno número de mutações pode conferir **novas propriedades químicas às proteínas virais**, resultando em mudanças na forma como o vírus se comporta nas infecções.

Mais informações sobre a pesquisa dos dados podem ser acessados na [RBAC](https://www.rbac.org.br/artigos/principais-variantes-do-sars-cov-2-notificadas-no-brasil/), revista brasileira de análises clínicas. Já os dados para o projeto são acessados no GitHub de **[Owid Covid19](https://github.com/owid/covid-19-data)**.

Para este projeto, realizaremos uma análise e verificaremos o panorama **mundial** ocorrido entre **janeiro de 2020 e janeiro de 2024**.

Os sintomas da COVID-19 podem variar de um **resfriado** a uma **síndrome gripal** - presença de um quadro respiratório agudo caracterizado por pelo menos dois dos seguintes sintomas: **sensação febril ou febre** associada a dor de garganta, dor de cabeça, tosse, coriza, ou até uma **pneumonia severa**. Sendo assim, os sintomas mais comuns são:

* Tosse;
* Febre;
* Coriza;
* Dor de garganta;
* Dificuldade para respirar;
* Perda de olfato (anosmia);
* Alteração do paladar (ageusia);
* Distúrbios gastrintestinais (náuseas/vômitos/diarreia);
* Cansaço (astenia);
* Diminuição do apetite (hiporexia);
* Dispnéia ( falta de ar).

A **transmissão** acontece de uma pessoa doente para outra ou por contato próximo por meio de:

* Toque do aperto de mãos contaminadas;
* Gotículas de saliva;
* Espirro;
* Tosse;
* Catarro;
* Objetos ou superfícies contaminadas, como celulares, mesas, talheres, maçanetas, brinquedos, teclados de computador etc.

Informações constantes no portal do [ministério da saúde](https://coronavirus.saude.gov.br/sobre-a-doenca).

Criaremos **cópias** da base de dados para análises mais assertivas: tanto para o nível **mundial**, quanto para os continentes (**Europa, Ásia, África, América do Sul, e Oceania**), assim como para os **países** em geral e uma para o **Brasil**. Exploraremos algumas informações relevantes como a linha do tempo para o **total de casos**. Para os países, o **total de mortes** e **total de mortes por 1 milhão de pessoas**.

```
# 14 - Criando cópias da base de dados com filtros específicos
df_copia = df.copy()
df_mundo = df_copia[(df_copia.location == "World") & (df_copia.date <= "2024-01-01")]
df_europa = df_copia[(df_copia.continent == "Europe") & (df_copia.date <= "2024-01-01")]
df_asia = df_copia[(df_copia.continent == "Asia") & (df_copia.date <= "2024-01-01")]
df_africa = df_copia[(df_copia.continent == "Africa") & (df_copia.date <= "2024-01-01")]
df_america_sul = df_copia[(df_copia.continent == "South America") & (df_copia.date <= "2024-01-01")]
df_oceania = df_copia[(df_copia.continent == "Oceania") & (df_copia.date <= "2024-01-01")]
df_brasil = df_copia[(df_copia.location == "Brazil") & (df_copia.date <= "2024-01-01")]
df_paises = df_copia.query("location not in ['United Kingdom', 'Asia', 'Africa', 'Europe', 'European Union', 'International', 'North America', 'Oceania', 'South America', 'World']")
df_paises = df_paises.loc[df_paises.date <= "2024-01-01"]
```

Visualizando a linha do tempo de **total de casos no mundo**.

```
# 15 - Visualizando o total de casos por COVID no mundo
dia = df_mundo.groupby("date")["total_cases"].sum()
dia = pd.DataFrame(dia)
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize = (10 ,3), dpi = 150)
dia["total_cases"].plot(kind = "line", ax = ax, lw = 2.5)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: format(int(x), ",")))
ax.set_title("Linha do tempo do total de casos de COVID-19 no mundo até Jan/24", fontsize = 14, fontweight = "semibold", color = "grey")
ax.set_xlabel("")
plt.tight_layout()
```

Olhando para o gráfico acima percebemos um **nítido crescimento** de casos próximo ao final de **2020** indo praticamente até **janeiro de 2022**. Em seguida, um **crescimento acentuado** entre **janeiro de 2022 e janeiro de 2023**, com uma relativa **estabilização** a partir desta data.

Com essa informação, vamos analisar os tópicos: **total de mortes** e **total de mortes a cada 1 milhão de pessoas** pelos **principais países** e continentes.

```
# 16 - Visualizando o total de mortes por país no mundo
ultima_data = df_paises['date'].max().strftime("%Y-%m-%d")
paises_ordenados = df_paises.loc[df_paises.date == ultima_data].sort_values(by = "total_deaths", ascending = False)
paises_ordenados = paises_ordenados[~paises_ordenados['location'].str.contains('income')][0:7]
paises_ordenados = paises_ordenados[["location", "total_deaths"]]

sns.set_style()
fig, ax = plt.subplots(figsize=(10, 3), dpi = 150)
barras = sns.barplot(x = "total_deaths", y = "location", data = paises_ordenados, palette = "cividis",
                     hue = "location", legend = False)

for i in barras.patches:
  ax.annotate(int(i.get_width()),
              xy = ((i.get_width() - (i.get_height() * 5000)),
                     i.get_x() + i.get_y() + (i.get_height() / 2)),
                     va = "center_baseline", color = "white", fontweight = "bold",
                     ha = "right")

for i in range(len(paises_ordenados)):
  pais = paises_ordenados.iloc[i, 0]
  valor = paises_ordenados.iloc[i, 1]
  ax.text(valor - valor + 5000, i, pais, fontsize = 10, fontweight = "bold", color = "white", va = "center")

plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
ax.text(0, -0.8, "Total de mortes por país no mundo até Jan/24", fontsize = 14, color = "grey", fontweight = "bold")
ax.set_xlabel("")
ax.set_ylabel("")
plt.tight_layout()
```

Acima são os dados dos países onde houve **maior número de mortes** por **COVID-19**. Em seguida, vamos visualizar o **total de mortes** a cada **1 milhão** de pessoas por países no **mundo**.

```
# 17 - Visualizando o total de mortes por país no mundo a cada 1 milhão de pessoas
ultima_data = df_paises["date"].max().strftime("%Y-%m-%d")
paises_ordenados = df_paises.loc[df_paises.date == ultima_data].sort_values(by = "total_deaths_per_million", ascending = False)
paises_ordenados = paises_ordenados[~paises_ordenados["location"].str.contains("income")][0:7]
paises_ordenados = paises_ordenados[["location", "total_deaths_per_million"]]

sns.set_style()
fig, ax = plt.subplots(figsize=(10, 3), dpi = 150)
barras = sns.barplot(x = "total_deaths_per_million", y = "location", data = paises_ordenados, palette = "copper",
                     hue = "location", legend = False)

for i in barras.patches:
  ax.annotate(int(i.get_width()),
              xy = ((i.get_width() - (i.get_height() * 30)),
                     i.get_x() + i.get_y() + (i.get_height() / 2)),
                     va = "center_baseline", color = "white", fontweight = "bold",
                     ha = "right")

for i in range(len(paises_ordenados)):
  pais = paises_ordenados.iloc[i, 0]
  valor = paises_ordenados.iloc[i, 1]
  ax.text(valor - valor + 30, i, pais, fontsize = 10, fontweight = "bold", color = "white", va = "center")

plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
ax.text(0, -0.8, "Total de mortes por país no mundo a cada 1 milhão de pessoas até Jan/24",
        fontsize = 14, color = "brown", fontweight = "bold")
ax.set_xlabel("")
ax.set_ylabel("")
plt.tight_layout()
```

Interessante notar que dificilmente países como **Peru, Bulgária ou Bósnia** seriam colocados como os locais com as **maiores quantidades de mortes** quando se trata de **mortes por 1 milhão de pessoas**.
